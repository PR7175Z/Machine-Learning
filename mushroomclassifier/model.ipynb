```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
dataset = pd.read_csv('data/mushrooms.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
display(dataset)
```

```python
x = dataset.iloc[:, 1:23]

label_encoder = LabelEncoder()

for i in x:
    x[i] = label_encoder.fit_transform(x[i])

y = dataset.iloc[:, 0]

print(x)
print(y)
```

```python
xtrain, xvalidate, ytrain, yvalidate = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(xtrain)
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(xtrain, ytrain)
pred = dtree.predict(xvalidate)
```

```python
from sklearn.metrics import accuracy_score,  confusion_matrix, classification_report
```

```python
print(accuracy_score(yvalidate, pred))
```

```python
print(classification_report(yvalidate, pred))
```

```python
print(confusion_matrix(yvalidate, pred))
```

```python
import pickle

with open('model.pkl', 'wb') as f:
    pickle.dump(dtree, f)
```
