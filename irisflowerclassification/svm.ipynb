```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
iris = load_iris()

datasets = pd.DataFrame(iris.data, columns=iris.feature_names)
datasets['target'] = iris.target
datasets['species'] = datasets['target'].apply(lambda x: iris.target_names[x])
```

```python
x = datasets.drop(['target', 'species'], axis=1)
y = datasets['species']

display(x, y)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python

#make_pipeline : function is used to create a pipeline that sequentially applies a list of transformers and a final estimator
model = make_pipeline(StandardScaler(), SVC())

# #without using make_pipeline:
# scaler = StandardScaler()
# x_reshape = scaler.fit_transform(x)
# y_reshape = scaler.transform(y)

# svm = SVC()


xtrain, xvalidate, ytrain, yvalidate = train_test_split(x, y, test_size=0.2, random_state=42)
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xvalidate)
display(ypred)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
print(confusion_matrix(yvalidate, ypred))
print(accuracy_score(yvalidate, ypred))
print(classification_report(yvalidate, ypred))
```
